```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=(100000, 1))
uniform_data = np.random.uniform(low=-2.5, high=2.5, size=(100000, 1))
normal_data = np.random.normal(loc=0, scale=1, size=(100000, 1))

plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")

plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/IT21034268/deep-learning-lab-repository/deep-learning-lab-01/Au_nanoparticle_dataset.csv')

# Define the features
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Calculate the Pearson and Spearman correlation
pearson_corr = df[features].corr(method='pearson')
spearman_corr = df[features].corr(method='spearman')

# Visualize the results using heatmaps
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title('Spearman Correlation')

plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/IT21034268/deep-learning-lab-repository/deep-learning-lab-01/Au_nanoparticle_dataset.csv')

filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
print(filtered_df.head(20))
```

```python
print(filtered_df.head(20))
```

```python
stats = filtered_df.describe()
print(stats)
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.hist(filtered_df['N_total'], bins=30, color='blue', alpha=0.7)
plt.title('N_total')

plt.subplot(1, 4, 2)
plt.hist(filtered_df['N_bulk'], bins=30, color='green', alpha=0.7)
plt.title('N_bulk')

plt.subplot(1, 4, 3)
plt.hist(filtered_df['N_surface'], bins=30, color='red', alpha=0.7)
plt.title('N_surface')

plt.subplot(1, 4, 4)
plt.hist(filtered_df['R_avg'], bins=30, color='purple', alpha=0.7)
plt.title('R_avg')

plt.show()
```

```python
sns.pairplot(filtered_df)
plt.show()
```

```python
g = sns.PairGrid(filtered_df)
g.map_upper(sns.histplot)  # Bivariate histogram
g.map_diag(sns.histplot, kde=True)  # Histogram with KDE
g.map_lower(sns.kdeplot)  # Bivariate KDE
plt.show()
```
